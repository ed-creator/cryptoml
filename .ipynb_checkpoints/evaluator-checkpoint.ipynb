```python
#a class to evaluate the various outcomes
class Evaluator():
    
    def __init__(self, data=preprocessing.DataProcessing()):
        self.data = data #
        self.erc20_names = self.data.erc20_names
        self.models = {}
        self.performance_data = {}
        
    #load a pre-built model into the class    
    def load_model(self,model_file, name):        
        self.models[name] = load_model('my_model.h5')
        
    #run model on a single dataset    
    def run_model(self,model_name, erc20):
        model = self.models[model_name]
        targets = evaluator.data.eval_sets[erc20]['eval_y']
        features = evaluator.data.eval_sets[erc20]['eval_X']
        pred_proba = model.predict(evaluator.data.eval_sets[erc20]['eval_X']).flatten()
        pred = pred_proba.round()
        targets = evaluator.data.eval_sets[erc20]['eval_y']
        print(model.evaluate(evaluator.data.eval_sets[erc20]['eval_X'], 
                             evaluator.data.eval_sets[erc20]['eval_y'], 
                             verbose=1))
        return pred_proba, pred, targets
        
        
```

```python
evaluator = Evaluator()
evaluator.load_model('my_model.h5', 'lstm_v1')
evaluator.models['lstm_v1']
p = evaluator.run_model('lstm_v1', 'gnt')
```
