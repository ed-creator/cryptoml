```python

# import ipynb.fs.full.test
# from ipynb.fs.full.test import hello
import ipynb.fs.defs.crypto
```

```python
import numpy as np
import pandas as pd
```

```python
DataProcessing = ipynb.fs.defs.crypto.DataProcessing()
```

```python
#Create list of ERC20 tokens to be evaluated
erc20_sets=['ae','bat','fun','gno','gnt','loom','omg','loom','omg','rep','salt','snt','zrx']

for erc20 in erc20_sets:
    csv = erc20 + '.csv'
    DataProcessing.process_raw_dataset(csv, erc20)
```

```python
DataProcessing.set_train_test_values(DataProcessing.final_sets.get('rep'))
```

```python
DataProcessing.set_eval_values()
```

```python
DataProcessing.build_model()
```

```python
DataProcessing.fit_model()
```

```python
scores = DataProcessing.model.evaluate(DataProcessing.eval_sets['ae']['test_X'], DataProcessing.eval_sets['ae']['test_y'], verbose=1)
```

```python
print("Accuracy: %.2f%%" % (scores[1]*100))
```

```python
predictions = pd.DataFrame()
labels = pd.DataFrame()
for key in DataProcessing.final_sets:
    predictions[key] = pd.Series([])
    preds = DataProcessing.model.predict(DataProcessing.eval_sets[key]['test_X'])
#     import pdb; pdb.set_trace()
    predictions[key] = preds.flatten()
    labels[key + '_y'] = DataProcessing.eval_sets[key]['test_y']
```

```python
# predictions['gno'][0] = predictions['gno'][0][0]
predictions.head()
```

```python
labels['sum'] = labels.sum(axis=1)
labels.head()
```

```python
choices = predictions
for index, row in predictions.iterrows():
    third_largest = (row.nlargest(3)[2])
    for i, val in choices.iloc[index].iteritems():
        if val >= third_largest:
            choices.iloc[index][i] = True
        else:
            choices.iloc[index][i] = False
```
